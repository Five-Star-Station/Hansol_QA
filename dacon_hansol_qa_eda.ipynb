```python
train_label = train_df[['id']].copy()
keywords = ['장점', '단점', '원인', '대책', '종류', '방법', '방식']
cols = ['질문_1', '질문_2']
for col in cols:
    train_label[col] = [list() for _ in range(train_df.shape[0])]
    for kw in keywords:
        train_label.loc[train_df[col].str.contains(kw), col] = train_label.loc[train_df[col].str.contains(kw), col].map(lambda x: x +[kw])
```
