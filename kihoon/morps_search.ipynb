```python
import pandas as pd
import numpy as np
from konlpy.tag import Hannanum

HAN = Hannanum()

def to_morphs(text):
  morps = HAN.morphs(text)
  return morps

def to_text(morps):
  text = ' '.join(morps)
  return text

def process(text):
  morps = to_morphs(text)
  text = to_text(morps)
  return text
```

```python
test_csv = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dacon/도배하자_QA/test.csv')
train_csv = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dacon/도배하자_QA/train.csv')
```

```python
test_questions = test_csv['질문'].apply(process)

print(test_questions.to_list())
train_questions = train_csv['질문_1'].to_list() + train_csv['질문_2'].to_list()
train_questions = list(map(process, train_questions))
train_questions
```

```python
## tf idf - case 1 질문만 TF-IDF vector로 이용하는 경우
from copy import deepcopy

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

from collections import deque


# documents = deepcopy(document2.insert(0, document1))

# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_vectorizer.fit(train_questions)
def find_most_close_one(query, documents):

  answers = tfidf_vectorizer.transform(documents)
  query_vec = tfidf_vectorizer.transform([query])

  answers_samples, answers_features = answers.shape

  most_close_one = None
  most_close_one_score = 0000

  for i in range(answers_samples):
    if cosine_similarity(query_vec, answers[i]) >= most_close_one_score:
      most_close_one = i
      most_close_one_score = cosine_similarity(query_vec, answers[i])

  return documents[most_close_one], most_close_one_score

def find_top_k_most_close_one(query, documents, k=3):

  # tf - idf로 벡터화
  answers = tfidf_vectorizer.transform(documents)
  query_vec = tfidf_vectorizer.transform([query])


  answers_samples, answers_features = answers.shape

  most_close_one = None
  most_close_one_score = -9999

  #queue
  queue = deque(((0,0),(0,0),(0,0)),maxlen=k)
  # queue.appendleft(most_close_one_score)

  #find top-k closest ones
  for i in range(answers_samples):
    similarity = cosine_similarity(query_vec, answers[i])

    if similarity >= most_close_one_score:
      most_close_one_score = similarity
      queue.appendleft((documents[i],most_close_one_score))
    elif similarity <= most_close_one_score and queue[1][1] <= similarity:
      queue.pop()
      queue.insert(1, (documents[i], similarity))
    elif queue[2][1] <= similarity and queue[1][1] >= similarity:
      queue.pop()
      queue.append((documents[i], similarity))


  return list(queue)





```

```python
print(test_questions[0])
find_most_close_one(test_questions[0], train_questions)
```

```python
print(test_questions[71])
find_top_k_most_close_one(test_questions[71], train_questions, k = 3)
```

```python
from tqdm import tqdm

records = []
for i in tqdm(range(len(test_questions))):
    query = [test_questions[i]]
    query_result = find_top_k_most_close_one(test_questions[i], train_questions, k = 3)
    questions, scores = zip(*query_result)
    record = query + list(questions) + list(scores)
    records.append(record)


records_df = pd.DataFrame.from_records(
    records,
    columns = [
        'test_questions',
        'train_questions_1',
        'train_questions_2',
        'train_questions_3',
        'train_scores_1',
        'train_scores_2',
        'train_scores_3',
        ])
records_df.to_csv('/content/drive/MyDrive/Colab Notebooks/dacon/도배하자_QA/morps_test_train_similarity_.csv')
```
